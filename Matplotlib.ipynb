```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0, 6, 9])
ypoints = np.array([5, 10, 20])

# draws a line with points
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# draws only points
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
ypoints = np.array([1, 8, 4, 10, 3])

# without x points defaults are 0, 1..
plt.plot(ypoints)
plt.show()
```

```python
# marker adds points to the graph
plt.plot(ypoints, marker = 'D')
plt.show()
```

```python
# display notation fmt (marker|line|color)
plt.plot(ypoints, 'p--m')
plt.show()
```

```python
# markersize parameter ms
# marker edge color mec
# marker face colore mfc
plt.plot(ypoints, marker = '*', ms = 20, mec = 'r', mfc = 'g')
plt.show()
```

```python
# line style parameter ls/linestyle
# line color parameter color
# line width parameter lw/linewidth
plt.plot(ypoints, marker = '*', ms = 15, ls = ':', color = 'pink', lw = 20)
plt.show()
```

```python
# to draw several lines just call plt.plot several times
ypoints1 = np.array([1,6, 5, 1, 9])
ypoints2 = np.array([7, 4, 1, 6, 3])

plt.plot(ypoints1)
plt.plot(ypoints2)
plt.show()
```

```python
# or in one call

xpoints1 = np.array([0, 1, 2, 3, 4])
xpoints2 = np.array([0, 1, 2, 3, 4])

plt.plot(xpoints1, ypoints1, xpoints2, ypoints2)
plt.show()
```

```python
ypoints = np.array([7, 4, 1, 6, 3])

plt.plot(ypoints)

# setting fonts
font1 = {'family': 'serif', 'color': 'blue', 'size': 30}
font2 = {'family': 'serif', 'color': 'red', 'size': 50}

# Title and axis
# fontdict to set font
# loc to locate title
plt.title("Title", fontdict = font2, loc = 'left')
plt.xlabel("X axis", fontdict = font1)
plt.ylabel("Y axis", fontdict = font1)

plt.show()
```

```python
ypoints = np.array([7, 4, 1, 6, 3])

plt.plot(ypoints)

# shows grid
# axis - if want only one axis
# same properties for style as for line
plt.grid(axis = 'x', color = 'green', ls = '--', lw = 2)

plt.show()
```

```python
# subplots has 3 params: n of rows, n of columns, n of the plot
#

#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
# title for subplot
plt.title('plot 1')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('plot 2')

plt.suptitle('Big title')
plt.show()
```

```python
# scatter plots
# plot 1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# attribute for color
plt.scatter(x, y, color = 'r')

# plot 2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)


plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]) * 7

# to color each dot separately use only c 
# cmap for mapping colors
# s for sizes
# alpha for transparency
plt.scatter(x, y, c=colors, cmap = 'viridis', s = sizes, alpha = 0.8)

# to show bar
plt.colorbar()

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# for bar chart
plt.bar(x,y, width = 0.4)
plt.show()
```

```python
# for horizontal representation
# color for color
# height/width for bars width
plt.barh(x, y, color = 'pink', height = 0.4)
plt.show()
```

```python
# creates normal distribution around 170 with std 10 and 250 data points
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mlabels = ['R', 'G', 'O', 'B']
mexplode = [0.2, 0.3, 0, 0]
mcolors = ['r', 'g', 'y', 'b']

# labels anticlockwise
# startangle - where data begins, default 0, at 3 o'clock
# explode - how far from centre are pies

plt.pie(y, labels = mlabels, startangle = 90, explode = mexplode, colors = mcolors)
# to add legend
plt.legend(title = 'Legend title')
plt.show() 
```
